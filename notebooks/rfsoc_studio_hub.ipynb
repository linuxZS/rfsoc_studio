## Featured Projects <a class="anchor" id="featured_projects"></a>
There are two featured RFSoC tools; the spectrum analyser and the RFSoC frequency planner. The spectrum analyser and frequency planner was developed by the University of Strathclyde Software Defined Radio (SDR) research laboratory. You can use the spectrum analyzer tool to explore your surrounding RF spectrum. The frequency planner tool was based on the original RFSoC frequency planner found [here](https://www.xilinx.com/publications/products/tools/zynq-ultrascale-plus-rfsoc-frequency-planner-rev1p7.xlsx.zip). You can use this to plan your RFSoC radio system design.

## Learn DSP for Communications <a class="anchor" id="learn_dsp_for_communications"></a>

The material consists of Jupyter notebooks that can be run on Jupyter on your computer, or on any PYNQ enabled board. These notebooks present introductory material for the following.
